```python
def datetime_translate(date):
    date, month, year = date.split('/')
    return '{Year}-{Month}-{Date}'.format(Year=year, Month=month, Date=date )

def hour_to_second(time):
    hour, minute = time.split(' ')[1].split(':')
    return int(hour)*60 + int(minute)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace0 = go.Scatter(
    x=[datetime_translate(new_data.datetime[i].split(',')[1]) for i in range (146)],
    y=[hour_to_second(new_data.datetime[i].split(',')[2]) for i in range (146)],
    text=new_data.title,
    mode='markers',
    hoverinfo = 'text',
    marker=dict(
        #color=['rgb(93, 164, 214)', 'rgb(255, 144, 14)',
               #'rgb(44, 160, 101)', 'rgb(255, 65, 54)']*20,
        color = 'rgb(93, 164, 214)',
        opacity=[1, 0.8, 0.6, 0.4],
        size = new_data.num_words,
        sizemode = 'area',
        sizeref = 1
    )
)

layout = go.Layout(
    title = "Freqency of articles released about Thailand rescue mission from June 26th to July 13th by dates",
    xaxis = dict(
        range = ['2018-6-26','2018-7-13']),
    yaxis = dict( 
        range = [0, 1490])
)

plot_data = [trace0]

fig = dict(data=plot_data, layout=layout)
py.iplot(fig, filename='Freqency of articles released about Thailand rescue mission from June 26th to July 13th by dates')
```

```python
DATE_IN_WEEK = {
    'Thứ hai' : 'Monday', 
    'Thứ ba'  : 'Tuesday',
    'Thứ tư'  : 'Wednesday',
    'Thứ năm' : 'Thursday',
    'Thứ sáu' : 'Friday',
    'Thứ bảy' : 'Saturday',
    'Chủ nhật': 'Sunday'
}

WEEKDAY_DATA_CONTAINER = {
    'Monday'   : 0,
    'Tuesday'  : 0,
    'Wednesday': 0,
    'Thursday' : 0,
    'Friday'   : 0,
    'Saturday' : 0,
    'Sunday'   : 0,
}
raw_weekday_data = dict(new_data.groupby(['weekday']).count().id)

for aday in raw_weekday_data.keys():
    WEEKDAY_DATA_CONTAINER[DATE_IN_WEEK[aday]] = raw_weekday_data[aday]



weekday_data = [go.Bar(
            x=list(WEEKDAY_DATA_CONTAINER.keys()),
            y=list(WEEKDAY_DATA_CONTAINER.values())
    )]

py.iplot(weekday_data, filename='Articles released about Thailand rescue mission from June 26th to July 13th by weekdates')
```
